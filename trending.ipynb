```
%matplotlib inline
plt.figure(figsize=(35,25))
sns.barplot(x=data['Crop'], y=data['Production']);
```

```
state = data.groupby("Crop")[["Area","Production"]].sum().sort_values(by = "Production", ascending = True)
plt.figure(figsize = (20,20))
state[:25].plot(kind = "bar",color = ["k","red"],edgecolor = "#000000")
plt.title("ARea and Production of top 25 States")
plt.xlabel("Crop")
plt.ylabel("Area/Production")
plt.grid(True)
```

```
state = data.groupby("Season")[["Area","Production"]].sum().sort_values(by = "Production", ascending = True)
plt.figure(figsize = (20,20))
state[:25].plot(kind = "bar",color = ["k","blue"],edgecolor = "#000000")
plt.title("ARea and Production of top 25 States")
plt.xlabel("Crop")
plt.ylabel("Area/Production")
plt.grid(True)
```
